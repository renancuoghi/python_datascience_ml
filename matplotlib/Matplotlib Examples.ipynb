```python
import matplotlib .pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x =  np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTION METHOD
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

## PART II

```python
fig, axes = plt.subplots(nrows=1,ncols=1)

#for current_ax in axes:
#    current_ax.plot(x,y)
axes.plot(x,y)
axes.set_title('First plot')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

#for current_ax in axes:
#    current_ax.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('Second plot')
```

## Figure size and dpi

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```

## ADD LEGEND

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Square')
ax.plot(x,x**3,label='X Cubed')
#ax.legend()
ax.legend(loc=0) # 0 = best matplot find
#ax.legend(loc=(0.1,0.1)) #define position manually
```

## PLOT APPEARANCE

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',lw=3,alpha=1,ls='--',marker='o',
        markersize=10,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green') #RGB Hex Code
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',lw=2,ls='--') #RGB Hex Code
# these do zoom on this limit values
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
