```python
import seaborn as sns
import numpy as np
%matplotlib inline
iris= sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
#g.map(plt.scatter)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
